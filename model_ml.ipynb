```python
## apply SVC model
from sklearn.svm import SVC
classifier_scv = SVC(gamma = 'scale', kernel = 'linear')
app_model(classifier_scv)
```

```python
## apply decision tree model
from sklearn.tree import DecisionTreeClassifier
classifier_dtc = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
app_model(classifier_dtc)
```

```python
## apply random forest model
from sklearn.ensemble import RandomForestClassifier
classifier_rfc = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
app_model(classifier_rfc)
```

```python
## apply k-nearest-neighbors model
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
app_model(classifier_knn)
```

```python
## apply AdaBoost model
from sklearn.ensemble import AdaBoostClassifier
classifier_abc = AdaBoostClassifier()
app_model(classifier_abc)
```
