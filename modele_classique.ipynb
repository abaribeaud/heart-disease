```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

df = pd.read_csv('heart_clean.csv').iloc[:, 1:]
```

```python
## Apply a model depending of the classifier
def app_model(classifier):
    
    X = df.drop(['malade'], axis = 1)
    y = df['malade']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

    classifier.fit(X_train, y_train)

    y_pred = classifier.predict(X_test)
    print("\nClassifier name : %s" % classifier)
    print("Accuracy : %s" % accuracy_score(y_test, y_pred))
    print("Confusion Matrix : \n", confusion_matrix(y_test, y_pred))
    print("Classification Report : \n", classification_report(y_test, y_pred))
```

## Apprentissage supervis√©e

```python
## apply SVC model
from sklearn.svm import SVC
classifier_scv = SVC(gamma = 'scale', kernel = 'linear')
app_model(classifier_scv)
```

```python
## apply k-nearest-neighbors model
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
app_model(classifier_knn)
```

```python
## apply k-nearest-neighbors model
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 25, metric = 'minkowski', p = 2)
app_model(classifier_knn)
```

```python
## apply k-nearest-neighbors model
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 50, metric = 'minkowski', p = 2)
app_model(classifier_knn)
```

```python
## apply k-nearest-neighbors model
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors = 100, metric = 'minkowski', p = 2)
app_model(classifier_knn)
```

```python
## apply decision tree model
from sklearn.tree import DecisionTreeClassifier
classifier_dtc = DecisionTreeClassifier(criterion = 'entropy', random_state = 0, max_depth=10)
app_model(classifier_dtc)
from sklearn import tree
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(classifier_dtc, filled=True)
```

```python
classifier_dtc2 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0, max_depth=10, min_samples_leaf=100)
app_model(classifier_dtc2)
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(classifier_dtc2, filled=True)
```

```python
## apply logistic regression
from sklearn.linear_model import LogisticRegression
classifier_log = LogisticRegression(random_state = 0)
app_model(classifier_log)
plt.figure(figsize=(16,16))
#plt.bar(x = df.columns.values, height = np.append(classifierlog.intercept, classifierlog.coef[0]))
plt.bar(x = df.drop(['malade'], axis=1).columns.values, height = classifier_log.coef_[0])
plt.xticks(rotation= 80)
print("Intercept : %s" % classifier_log.intercept_)
```
