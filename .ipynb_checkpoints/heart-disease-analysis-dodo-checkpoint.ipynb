## Exploration données quantitatives

```python
label_quant = ["age", "taille", "Poids", "pression_systo", "pression_diasto"]

df_quant = df_heart[label_quant]
```

```python
def make_box_hist(target, df):
    """
        Création d'un Boxplot et d'un Historgramme
    """
    
    plt.figure(figsize=(16,8))

    plt.boxplot(df[target]);
    plt.title('Boxplot et Histograme de la variable %s' % target)
    plt.xlabel('Taille', fontsize=12)
    plt.axes([0.62,0.65,0.25,0.20], facecolor="#ffe5c1");
    plt.xticks([])
    plt.yticks([])
    plt.xlabel('Distribution', fontsize=12)
    plt.hist(df[target], bins=40, color="#FFC575");
```

```python
sns.set(rc={"figure.figsize": (20, 20)});

for i in range(0,5):
        make_box_hist(label_quant[i], df_heart)
```

### Valeurs abberantes  

On observe pour la pression diasto et systo des valeurs abberantes, il est donc nécessaire de tenter d'en apprendre d'avantage sur ces dernières :  

**Est-il possible d'avoir une tension diasto et systo de l'ordre du millier dans la vrai vie ?**  
Les personnes ayant une pression systo et dysto, respectivement supérerieur à, 180 et 110 sont déjà en hypertension très dangereuse.
Il est donc impossible d'aovoir une pression diastolique ou systo superiéeur à 1000 par exemple  
  
De la même façon, une pression négative est également impossible

```python
# Pression systo négative
display(df_quant[df_quant['pression_systo'] < 0])

# Pression diasto négative
display(df_quant[df_quant['pression_diasto'] < 0])
```

On observe très peu de valeur négative pour la pression, de plus, les valeurs absolues des pressions négatives ne semblent pas abbérante sur notre jeu de données.  
On décide donc de les garder en les normalisant avec la fonction `.abs()`

```python
df_quant["pression_diasto"] = df_quant["pression_diasto"].abs()
df_quant["pression_systo"] = df_quant["pression_systo"].abs()

df_quant.describe()
```

Il faut ensuite détecter les valeurs abberantes "impossible" telle qu'une pression a 0 ou à 1000 
On utilise la règle de détection des outputs en se basant sur l'écart inter-quartil : IQR

```python
# Première approche : The boxplot rule for outlier detection (Q3 + (1.5 * IQR))

def outlier_detection_box_1(target, df):
    print(target)
    q1 = df_heart[target].quantile(q=0.25)
    q3 = df_heart[target].quantile(q=0.75)
    med = df_heart[target].median()

    iqr = q3 - q1

    upper = q3 + (1.5 * iqr)
    lower = q1 - (1.5 * iqr)

    print("\tUpper : %s | shape : %s" % (upper, df_heart[df_heart[target] > upper].shape))
    print("\tLower : %s | shape : %s\n" % (upper, df_heart[df_heart[target] < lower].shape))

    return upper, lower
    
```

```python
outlier_detection_box_1("pression_systo", df_quant)
outlier_detection_box_1("pression_diasto", df_quant)
```

On perd encore trop de données qui pourraient être utiles. Une pression à 180 ou un peu plus est en effet largement possible.  

On change donc de technique de détection des valeurs abberantes

```python
# Deuxième approche : 

def outlier_detection_box_2(target, df):
    print(target)
    med = df_heart[target].median()
    std = df_heart[target].std()

    upper = med + 2*std
    lower = med - 2*std
    
    print("\tUpper : %s | shape : %s" % (upper, df_heart[df_heart[target] > upper].shape))
    print("\tLower : %s | shape : %s\n" % (lower, df_heart[df_heart[target] < lower].shape))
    
    return upper, lower
```

```python
upper_systo, lower_systo = outlier_detection_box_2("pression_systo", df_quant)
upper_diasto, lower_diasto = outlier_detection_box_2("pression_diasto", df_quant)
```

Ici, on s'assure à minima d'enlever les valeurs réelement abbarérantes : pression > 400 impossible

```python
df_quant = df_quant[df_quant["pression_systo"] < upper_systo]
df_quant = df_quant[df_quant["pression_systo"] > lower_systo]

df_quant = df_quant[df_quant["pression_diasto"] < upper_diasto]
df_quant = df_quant[df_quant["pression_diasto"] > lower_diasto]
```

```python
df_quant.describe()
```

On enlève ainsi les données totalement abberante ete impossible

```python
for i in range(0,5):
        make_box_hist(label_quant[i], df_quant)
```
